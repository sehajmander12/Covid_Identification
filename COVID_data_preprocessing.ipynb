```python
import pandas as pd
import os
import shutil
import random
import math
```

```python
file_path = "chestxray/metadata.csv"
images_path = "chestxray/images"

df = pd.read_csv(file_path)
print(df.shape)
```

```python
df.head()
```

```python
if not os.path.exists("dataset"):
    os.mkdir("dataset")

target_train = "dataset/train"
target_val = "dataset/val"

if not os.path.exists(target_train):
    os.mkdir(target_train)
    
if not os.path.exists(target_val):
    os.mkdir(target_val)
    
target_train_covid_p = "dataset/train/covid_p"
target_val_covid_p = "dataset/val/covid_p"

if not os.path.exists(target_train_covid_p):
    os.mkdir(target_train_covid_p)
    
if not os.path.exists(target_val_covid_p):
    os.mkdir(target_val_covid_p)
```

```python
count = 0

for(i,row) in df.iterrows():
    if row["RT_PCR_positive"] == "Y" and row["view"] == "PA":
        count += 1

print(count)

train = math.floor(count*0.8)
val = count - train

print(train,val)
```

```python
count = 0

for(i,col) in df.iterrows():
    
    if col["RT_PCR_positive"] == "Y" and col["view"] == "PA":
        file_name = col["filename"]
        image_path = os.path.join(images_path,file_name)
        if count < train:
            image_copy_path = os.path.join(target_train_covid_p,file_name)
        else:
            image_copy_path = os.path.join(target_val_covid_p,file_name)
            
        shutil.copy2(image_path,image_copy_path)
        count += 1
        print("Image ",count)
    
        
```

```python
kaggle_path = "chest_xray_kaggle/train/NORMAL"

target_train_normal = "dataset/train/normal"
target_val_normal = "dataset/val/normal"

if not os.path.exists(target_train_normal):
    os.mkdir(target_train_normal)

if not os.path.exists(target_val_normal):
    os.mkdir(target_val_normal)
```

```python
image_names = os.listdir(kaggle_path)
random.shuffle(image_names)
```

```python
for i in range(count):
    
    image_name = image_names[i]
    image_path = os.path.join(kaggle_path,image_name)
    
    if(i<train):
        target_path = os.path.join(target_train_normal,image_name)
    else:
        target_path = os.path.join(target_val_normal,image_name)
    
    shutil.copy2(image_path,target_path)
    print("Image ", i+1)
    
```
